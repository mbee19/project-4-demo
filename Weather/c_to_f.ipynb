```python
file_path = 'Resources/combined_disney_weather.csv'
df_file = pd.read_csv(file_path)
df = pd.DataFrame(df_file)
df.head()
```

```python
df['tavg_f'] = (df['tavg'] * 1.8) + 32

df.head()
```

```python
df['tmin_f'] = (df['tmin'] * 1.8) + 32

df.head()
```

```python
df['tmax_f'] = (df['tmax'] * 1.8) + 32

df.head()
```

```python
df.drop(['tavg', 'tmin', 'tmax'], axis = 1, inplace=True)
df.head()
```

```python
df.columns
```

```python
df = df[['time','tavg_f', 'tmin_f', 'tmax_f', 'prcp', 'snow', 'wdir', 'wspd', 'wpgt', 'pres', 'tsun',
       'Location']]

df.head()
```

```python
df.info()
```

```python
df.drop('tsun', axis=1, inplace=True)
df.head()
```

```python
df['date'] = pd.to_datetime(df['time'])
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df[['date', 'tavg_f','tmin_f', 'tmax_f', 'Location']]
df.head()
```
